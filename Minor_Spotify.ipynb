Find out and plot different clusters according to different parameters like playlist genres , playlist names.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
label_encoder_genre = LabelEncoder()
label_encoder_playlist = LabelEncoder()
```

```python
df['playlist_genre_encoded'] = label_encoder_genre.fit_transform(df['playlist_genre'])
df['playlist_name_encoded'] = label_encoder_playlist.fit_transform(df['playlist_name'])
```

```python
clustering_features = df[['track_popularity', 'tempo', 'loudness', 'valence', 'playlist_genre_encoded', 'playlist_name_encoded']]
```

```python
kmeans = KMeans(n_clusters=5, random_state=42) 
```

```python
df['cluster'] = kmeans.fit_predict(clustering_features)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tempo', y='track_popularity', hue='cluster', data=df, palette='Set1')
plt.title('Clusters based on Tempo and Track Popularity')
plt.show()
```
